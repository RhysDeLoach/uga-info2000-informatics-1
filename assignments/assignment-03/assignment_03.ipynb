```python
# Section 2: Math Operators
# Step 1: Assign variables and calculate give expressions
e = 2.71828
x = 2
y = 1.5
z = 4
expression_a = z*(1/2)/y+x
expression_b = (x+y)**2
expression_c = e**((-2*x)*(y+x))
print(f"Expression a is equal to {expression_a}")
print(f"Expression b is equal to {expression_b}")
print(f"Expression c is equal to {expression_c}")

# Step 2: BMI Calculator
height = input("Enter your height in inches:")
height = float(height)
height = height/39.37 # Converts to metric
weight = input("Enter your weight in pounds:")
weight = float(weight)
weight = weight/2.205 # Converts to metric
bmi = weight/(height**2)
print(f"Your BMI is {bmi}")

# Step 3: Ask for coefficients and solve for x using the quadratic equation
coefficients = input("Input the quadratic coefficients in order a,b,c separated by commas:")
print(coefficients)
a,b,c = coefficients.split(',')
a = float(a)
b = float(b)
c = float(c)
x = [(-b+(b**2-4*a*c)**(1/2))/2,(-b-(b**2-4*a*c)**(1/2))/2]
print(f"The roots of you quadratic function are x={x[0]} and x={x[1]}")
```

```python
# Section 3: Lists
# Step 1: Make a list with 10 numbers
my_list = [4,5,6,2,3,1,7,9,8,10]
print(f'Here is the list of ten numbers: {my_list}')

# Step 2: Print all elements except for the last
print(f'Here is the list of the first nine numbers: {my_list[:-1]}')

# Step 3: Store the 1st and 6th element as x and y, and find the square root of the sum
x = my_list[0]
y = my_list[5]
sum = x+y
squareroot = (sum)**(1/2)
print(f'x and y are {x} and {y} respectively, their sum is {sum}, and the root of their sum is {squareroot}')

# Step 4: Create empty list named list_2 and add x, y, and the root of the sum
list_2 = []
list_2.extend([x,y,sum,squareroot])

# Step 5: Print the list in reverse order
print(f'This is the reverse order of list_2: {sorted(list_2,reverse=True)}')

# Step 6: Extract the odd-numbered indexes of my_list and put them in a new list named odd_indexed_list
odd_indexed_list = my_list[1::2]
print(f'Here is the odd indexed list: {odd_indexed_list}')

# Step 7: Sort the odd indexed list
print(f'This is the odd indexed list sorted: {sorted(odd_indexed_list)}')

# Step 8: Extend the odd indexed list and print it's length
odd_indexed_list.extend([2,3,4])
print(f'Here is the odd indexed list extended: {odd_indexed_list}')
print(f'The length of the extended odd indexed list is {len(odd_indexed_list)}')

# Step 9: Delete my_list
del my_list
```

```python
# Section 4: Random Library
# Step 1: Import random library
import random

# Step 2: Create a list of 10 pizza toppings and use the random.sample() function to pick three
pizza_toppings = ["pepperoni", "mushrooms", "pineapple", "bacon", "ham", "chicken", "olives", "banana peppers", "onions", "sausage"]
print(f'The three selected toppings are {random.sample(pizza_toppings, 3)}')

# Step 3: Create a list of 5 local restaurants with their pricing and use the random.shuffle() function to pick one
local_restaurants = ["Little Italy $", "The Place $$", "Cali N Tito's $", "Dos Palmas $$", "Porterhouse Grill $$$"]
random.shuffle(local_restaurants)
selected_restaurant = local_restaurants[0]
print(f'The selected restaurant is {selected_restaurant}')

# Step 4: Create a list of 10 outdoor activities and use the random.choice() function to pick one
outdoor_activities = ["football", "soccer", "basketball", "hiking", "hunting", "fishing", "skiing", "shotput", "discus", "long jump"]
print(f'The selected outdoor activity is {random.choice(outdoor_activities)}')
```
