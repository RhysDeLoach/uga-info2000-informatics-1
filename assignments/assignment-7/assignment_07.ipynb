```python
# Import Libraries
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
# Section 1: Basic Plotting
x = np.array(range(-20,20))
y = 4*x**2+2
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Simple Line Plot')
plt.plot(x,y)
plt.show()
```

```python
# Section 2: Scatter Plot
x = np.random.rand(100,1)*100
y = np.random.rand(100,1)*100
plt.scatter(x,y, marker = 'o', edgecolor = 'red', facecolor = 'blue', label = 'Random (x,y)')
plt.legend(loc = 0)
plt.show()
```

```python
# Section 3: Bar Chart
x = np.array(["LSU", "Ole Miss", "Georgia", "Missouri","Tennessee"])
y = np.array([45.29,41.67,40.14,33.86,33.5])

plt.text(0,45.29+0.5,'45.29 tp/g',ha='center')
plt.text(1,41.67+0.5,'41.67 tp/g',ha='center')
plt.text(2,40.14+0.5,'40.14 tp/g',ha='center')
plt.text(3,33.86+0.5,'33.86 tp/g',ha='center')
plt.text(4,33.5+0.5,'33.5 tp/g',ha='center')

plt.xlabel('Top five SEC college football teams')
plt.ylabel('Total points per game (tp/g)')
plt.title('SEC College Football Total Points per Game Rankings')

plt.bar(x,y, color = ['purple','blue','red','yellow','orange'])
plt.show()
```

```python
# Section 4: Subplots
x = np.array(range(-20,20))

plt.subplot(2,2,1)
y1 = 4100*x**2+2
plt.plot(x,y1, c = 'r', label = '4100x^2+2')
plt.title('Function 1')
plt.legend(loc = 0)
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.subplot(2,2,2)
y2 = 55*x**2+17
plt.plot(x,y2, c = 'b', label = '55x^2+17')
plt.title('Function 2')
plt.legend(loc = 0)
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.subplot(2,2,3)
y3 = -600*x**2+40
plt.plot(x,y3, c = 'g', label = '-600x^2+40')
plt.title('Function 3')
plt.legend(loc = 0)
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.subplot(2,2,4)
y4 = -x**2+1
plt.plot(x,y4, c = 'y', label = '-x^2+1')
plt.title('Function 4')
plt.legend(loc = 0)
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.suptitle("Variations of Quadratic Functions")
plt.tight_layout()
plt.show()
```
