```python
# Section 2: Functions and List Comprehensions
# Part 1: Odd Cube Root Function

def oddCuberoot(number):
    if number%2 != 0 and number>0:
         print(f'The cube root is {number**(1/3)}')
    elif number == 0:
        print('The number is zero')
    elif number%2 == 0:
        print('The number is even')
    else:
        print('The number is negative')
               
```

```python
# Section 2: Functions and List Comprehensions
# Part 3: Strings using List Comprehensions

my_string = "Will I succeed if I tried that trick?"

shortened_string = [words for words in my_string.split(' ') if len(tuple(words)) > 2]
lowercase_string = [words.casefold() for words in shortened_string]
modified_string = lowercase_string
print(f'The modified string in list form is {modified_string}')
```

```python
# Section 2: Functions and List Comprehensions
# Part 4: Unique List

def uniqueElements(list):
    unique_list = []
    for element in list:
        if element not in unique_list:
            unique_list.append(element)
    print(unique_list)
    
```

```python
# Section 2: Functions and List Comprehensions
# Part 5: Cost Functions

def total_hotel_cost(days):
    nightly_hotel_cost = 140
    hotel_cost = nightly_hotel_cost * days
    print(f'You will spend ${hotel_cost} on hotels')
    return(hotel_cost)

def total_airfare_cost(city):
    airfare_cost_data = {'Charlotte':183,'Tampa':220,'Pittsburgh':222,'Atlanta':475}
    airfare_cost = airfare_cost_data[city]
    print(f'You will spend ${airfare_cost} on airfare')
    return(airfare_cost)

def total_rental_cost(days):
    daily_car_rental_cost = 40
    rental_cost = days * daily_car_rental_cost
    print(f'You will spend ${rental_cost} on car rental')
    return(rental_cost)

def total_cost():
    days = input('Enter the number of days you will be staying: ')
    city = input('Enter the city that you will be flying into: ')
    spending_money = input('Enter the amount of money you will spend in a day: ')
    days = float(days)
    spending_money = float(spending_money)
    total_spending = spending_money * days
    print(f'You will spend ${total_spending} in spending money')
    entire_cost = total_hotel_cost(days) + total_airfare_cost(city) + total_rental_cost(days) + spending_money
    print(f'You will spend ${entire_cost} on your trip')
    return(entire_cost)

    
```

```python
# Section 2: Functions and List Comprehensions
# Part 6: Checkerboard Function

def checker_board(dimension,element1,element2):
    row1 = []
    row2 = []
    checker_board = []
    if element1 == element2:
        print('Invalid')
    else:
        while len(row1) < dimension and len(row2) < dimension:
            row1.append(element1)
            row2.append(element2)
            if len(row1) == dimension and len(row2) == dimension:
                break
            row1.append(element2)
            row2.append(element1)
        while len(checker_board) < dimension:
            checker_board.append(row1)
            if len(checker_board) == dimension:
                break
            checker_board.append(row2)
        print(checker_board)
    
```
